```python
import matplotlib.pyplot as plt
all_ages['Median'].plot(kind='hist', bins=30, width=1000)
plt.title("收入中位数分布")
plt.xlabel("收入中位数")
plt.ylabel("频数")
plt.show()
```

```python
all_ages['Median'].plot(kind='hist', bins=50, width=5000)
plt.title("收入中位数分布")
plt.xlabel("收入中位数")
plt.ylabel("频数")
plt.show()
```

```python
category_median = all_ages.groupby('Major_category')['Median'].median().sort_values(ascending=False)
print(category_median)
```

```python
stem_categories = ['Engineering', 'Computers & Mathematics', 'Physical Sciences']
all_ages['Is_STEM'] = all_ages['Major_category'].apply(lambda x: x in stem_categories)
median_overall = all_ages['Median'].median()
low_earning_stem = all_ages[(all_ages['Is_STEM']) & (all_ages['Median'] <= median_overall)]
print(low_earning_stem[['Major', 'Median', 'P25th', 'P75th']].sort_values(by='Median', ascending=False))
```
